## Learning portfolio: Visualizing greedy colouring strategies

We know that `networkx` implements all these fancy graph coloring algorithms, but what do they actually look like? How do they compare to each other? Do they usually give similar results?

It's hard to visualize these things so I'm gonna write a script to generate animations for these different strategies. First I'm gonna mess around with animations in this script then I'm gonna make it into a webpage

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
import matplotlib.animation as animation
```

```python
# G = nx.random_lobster(10, 0.7, 0.7)
G = nx.dense_gnm_random_graph(30, 70)
nx.draw(G, with_labels=True)
```

```python
coloring = nx.greedy_color(G)
mapping = [coloring[n] for n in range(len(G.nodes))]
nx.draw(G, with_labels=True, node_color=mapping, pos = nx.kamada_kawai_layout(G))
```

```python
ordering = nx.algorithms.coloring.strategy_largest_first(G, None)
fig,ax = plt.subplots()
for i in range(1, 5):
    # plt.figure(f"Coloring first {i} nodes")
    ax.clear()
    first_i = ordering[:i]
    mapping = [coloring[n] + 1 if n in first_i else 0 for n in range(len(G.nodes))]
    nx.draw(G, with_labels=True, node_color=mapping, pos = nx.kamada_kawai_layout(G))
    # display(nx.draw(G, with_labels=True, node_color=mapping, pos = nx.kamada_kawai_layout(G)))
    # break
    display(fig)
```

```python
ordering = list(nx.algorithms.coloring.strategy_largest_first(G, None))
fig,ax = plt.subplots()
fig.set_size_inches(6,6)
max_color= max(coloring.values())

def plot_first_i(i):
    print(f"frame={i}")
    # clear current plot
    ax.clear()

    # plot new coloring
    first_i = ordering[:i]
    mapping = [coloring[n] + 1 if n in first_i else 0 for n in range(len(G.nodes))]
    nx.draw_networkx(G, with_labels=True, node_color=mapping, pos = nx.kamada_kawai_layout(G), ax=ax, cmap="Spectral", vmin=0,  vmax=max_color)
```

```python
anim = animation.FuncAnimation(fig, plot_first_i, frames=len(G.nodes)+1)
anim.save("test.mp4", fps=4)
# for i in range(5):
#     plot_first_i(i)
```
